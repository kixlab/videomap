```python
types_df = pd.read_csv ('./data/final/types_df.csv')
categories_df = pd.read_csv ('./data/final/categories_df.csv')
sections_df = pd.read_csv ('./data/final/sections_df.csv')
```

```python
types_df = types_df.drop(labels='Unnamed: 0', axis=1)
categories_df = categories_df.drop(labels='Unnamed: 0', axis=1)
sections_df = sections_df.drop(labels='Unnamed: 0', axis=1)
types_df.head()
```

```python
# types
type_cols = types_df.columns.values.tolist()
type_cols.remove('time')

types_data = {}
for col in type_cols:
    types_data[col] = []
    for ind in range (len(types_df)):
        for x in range (types_df[col][ind]):
            types_data[col].append (types_df['time'][ind])

# categories
cat_cols = categories_df.columns.values.tolist()
cat_cols.remove('time')

cats_data = {}
for col in cat_cols:
    cats_data[col] = []
    for ind in range (len(categories_df)):
        for x in range (categories_df[col][ind]):
            cats_data[col].append (categories_df['time'][ind] * 1000)

# sections
# sec_cols = sections_df.columns.values.tolist()
# sec_cols.remove('time')

# secs_data = {}
# for col in sec_cols:
#     secs_data[col] = []
#     for ind in range (len(sections_df)):
#         for x in range (sections_df[col][ind]):
#             secs_data[col].append (sections_df['time'][ind])

types_data['opening'][:10]
```
